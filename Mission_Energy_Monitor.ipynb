```python
def analisar_missao(nome_cenario, temperatura, comunicacao, bateria, geracao_solar, consumo_modulos, status_modulo):
    saldo_energia = geracao_solar - consumo_modulos
    alertas = []
    acoes = []

    if temperatura > 80:
        alertas.append("Temperatura elevada")
        acoes.append("Ativar controle térmico do módulo")

    if comunicacao == "Instável":
        alertas.append("Comunicação instável")
        acoes.append("Priorizar canal de comunicação principal")

    if bateria < 25:
        alertas.append("Nível de bateria baixo")
        acoes.append("Ativar modo de economia de energia")

    if saldo_energia < 0:
        alertas.append("Consumo maior que a geração solar")
        acoes.append("Reduzir consumo dos módulos não essenciais")

    if status_modulo == "Crítico":
        alertas.append("Módulo em estado crítico")
        acoes.append("Enviar alerta para a central da missão")

    if len(alertas) == 0:
        status_geral = "NORMAL"
    elif len(alertas) <= 2:
        status_geral = "ATENÇÃO"
    else:
        status_geral = "CRÍTICO"

    print("=" * 55)
    print("MISSION ENERGY MONITOR")
    print("=" * 55)
    print(f"Cenário: {nome_cenario}")
    print("-" * 55)
    print(f"Temperatura: {temperatura} °C")
    print(f"Comunicação: {comunicacao}")
    print(f"Bateria: {bateria}%")
    print(f"Geração solar: {geracao_solar} W")
    print(f"Consumo dos módulos: {consumo_modulos} W")
    print(f"Saldo energético: {saldo_energia} W")
    print(f"Status do módulo: {status_modulo}")
    print("-" * 55)
    print(f"Status geral da missão: {status_geral}")

    print("\nALERTAS:")
    if len(alertas) == 0:
        print("- Nenhum alerta detectado")
    else:
        for alerta in alertas:
            print(f"- {alerta}")

    print("\nAÇÕES AUTOMÁTICAS:")
    if len(acoes) == 0:
        print("- Manter operação normal")
    else:
        for acao in acoes:
            print(f"- {acao}")

    print("=" * 55)
```

```python
analisar_missao(
    nome_cenario="Operação normal",
    temperatura=42,
    comunicacao="Estável",
    bateria=78,
    geracao_solar=850,
    consumo_modulos=620,
    status_modulo="Operacional"
)
```
